## Matrix Multiplication

In this notebook I will try to give some intuition into the operation which is a matrix multiplication

```python
%load_ext lab_black
%load_ext autoreload
%autoreload 2
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

First let's start with a vector $v$ which is $2$ in the $\hat{x}$ direction. Similarly our vector has length $1$ in the $\hat{y}$ direction

$$v = 2 \hat{x} + \hat{y}= \begin{pmatrix} 2 \\
1 \end{pmatrix} $$

This vector can be seen below

```python
X, Y = np.array([1, 0]), np.array([0, 1])
```

```python
v = 2 * X + 1 * Y
```

```python
fig, ax = plt.subplots()
ax.quiver(0, 0, *X, angles="xy", scale_units="xy", scale=1, color="r")
ax.quiver(0, 0, *Y, angles="xy", scale_units="xy", scale=1, color="b")
ax.quiver(0, 0, *v, angles="xy", scale_units="xy", scale=1)

ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
plt.show()
```

In the plot above we can see that our unit vectors are represented by the blue and red arrows. The black arrow represents the vector $v$

### Matrix Multiplication

Here we will run through an explanation of matrix multiplication. In this section I will first show the transformation of the unit vectors. Then show how these transformed unit vectors form a new basis which the vector $v$ can be thought of as in.

As an example we will take the our matrix as 
$$A= \begin{pmatrix} 1 & 1 \\
2 & -3 \end{pmatrix} $$

```python
A = np.array([[1, 1], [2, -3]])
```

Let's take a look at how this matrix transforms our unit vectors

$$ A X = X' $$ and
$$ A Y = Y'$$

```python
X_prime = A @ X
Y_prime = A @ Y
```

```python
fig, ax = plt.subplots()
ax.quiver(0, 0, *X_prime, angles="xy", scale_units="xy", scale=1, color="r")
ax.quiver(0, 0, *Y_prime, angles="xy", scale_units="xy", scale=1, color="b")
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
plt.show()
```

From this you can see that the columns in matrix $A$ represent the transformed unit vectors $X', Y'$

Since we have transformed our unit vectors we can find $v'$ easily as
$$v' = 2\hat{x'} + 1\hat{y'}$$

```python
v_prime = 2 * X_prime + 1 * Y_prime
```

```python
fig, ax = plt.subplots()
ax.quiver(0, 0, *v_prime, angles="xy", scale_units="xy", scale=1)
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
plt.show()
```

As a check, let's verify our slightly more involved calculation by just plotting $v' = Av$

```python
fig, ax = plt.subplots()
ax.quiver(0, 0, *(A @ v), angles="xy", scale_units="xy", scale=1)
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
plt.show()
```

Great.

So to conclude. A matrix multiplication can be viewed as a transformation of the basis vectors. Any vector which is multiplied by a matrix retains it's coefficients, but the basis vectors are adapted. The basis vectors are adjusted according to the columns in the matrix.
